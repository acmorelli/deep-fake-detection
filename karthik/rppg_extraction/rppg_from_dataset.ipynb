### Extract rPPG signal from dataset (FaceForensic++) and store as a CSV file

```python
# Reference:
# https://github.com/ondyari/FaceForensics
# https://kaldir.vc.in.tum.de/faceforensics_download_v4.py

# Command to download dataset:
# python faceforensics_download_v4.py output_dir/ --dataset FaceShifter --compression c0 --type videos --server EU2
# refer the above website to know the specific commands to download the dataset in required formats.
```

```python
from pathlib import Path

selector = 3
dataset_list = ['original',             #0
                'FaceShifter',          #1
                'FaceSwap',             #2
                'NeuralTextures']       #3

max_length = 15
# Define the folder path and file format (e.g., .csv)
main_path = 'dataset/FaceForensic++/' + dataset_list[selector]+'/c23/'
folder_path = Path(main_path+'videos/')
csv_folder_path = main_path +'sig_csv/'
```

```python

from pathlib import Path
from grgb_sig import grgb_est
import csv
import os  

file_formats = ['*.mp4', '*.avi']  # Include your desired file format

#Create directory if it doesn't exist
if not os.path.exists(csv_folder_path):
    os.makedirs(csv_folder_path)

csv_fps_fname = main_path + 'vid_fps.csv'
with open(csv_fps_fname, 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Filename', 'FPS'])

#To limit the video count to definite numbers
vid_count = 300
count = 0

# Iterate through all files in the folder that match the format
for file_format in file_formats:
    
    for file_path in folder_path.glob(file_format):
        file_path_str = str(file_path)
        file_name = file_path.stem
        if count < vid_count:
            grgb_signal, fps = grgb_est(file_path_str)
        
            
            #Save the filtered signal data to csv file in respective directory
            csv_fname = csv_folder_path + file_name + '.csv'
            with open(csv_fname, 'w', newline='') as csvfile:
                writer = csv.writer(csvfile)
                for value in grgb_signal:
                    writer.writerow([value])

            #Save the video fps to csv file in respective directory
            with open(csv_fps_fname, 'a', newline='') as csvfile:
                writer = csv.writer(csvfile)
                writer.writerow([file_name, fps])
        else:
            break
        #count+=1

        print(f"Processed file_name: {file_name}")
```
